```python
import numpy as np, pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
path = 'labelled_historical_modus_1_decimal_lat_long.csv'

df = pd.read_csv(path, index_col=0)

df.head()
```

```python
df.dtypes
```

```python
# Label fires as rows without NANs in the STAT_CAUSE_DESCR column
df.loc[df['STAT_CAUSE_DESCR'].isna() != True, 'isFire'] = 1
df.loc[df['STAT_CAUSE_DESCR'].isna() == True, 'isFire'] = 0


# build more time features
df['DATE'] = pd.to_datetime(df['DATE'])
df['WEEK'] = df['DATE'].dt.weekofyear
df['DAY'] = df['DATE'].dt.day

# drop excess columns
df.drop(columns = ['LAT_ROUND_1', 'LON_ROUND_1', 'STAT_CAUSE_DESCR', 'instrument', 'acq_date'], inplace=True)

df.head()
```

```python
df.to_csv(path)
```
