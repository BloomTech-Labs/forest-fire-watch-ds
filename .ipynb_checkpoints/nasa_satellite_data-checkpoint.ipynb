```python
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    
    # haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 3956 #radius of earth in miles mean of  poles and equator radius
    return c * r
    
    
```

```python
# test mapping distances to fires in the modus df

# my actual cordinates 
# Latitude:47.653278°
# Longitude:-122.347204°

my_coords = [-122.347204, 47.653278]

def check_fires(user_coords, fire_cords, perimiter):
    """
    checks a single user long/lat against an array of long/lat fire coordinates
    and returns a dict of fires within the perimiter and a binary Alert: yes/no
    """
    
    results = {
        'Alert' : False,
        'Fires' : []
    }
    
    #iterate through the fire cord pairs
    for cord in fire_cords:
        
        #compare user_coords with individual fires
        distance = haversine(user_coords[0], user_coords[1], cord[0], cord[1])
        if distance <= perimiter:
            results['Alert'] = True
            
            fire_location_tuple = ((cord[0], cord[1]), distance)
            results['Fires'].append(fire_location_tuple)
        
        else:
            pass
    
    return results
```

```python
# test 1

fires = m_df[['longitude', 'latitude']].values
perimeter = 500

my_fires = check_fires(my_coords, fires, perimeter)
```

```python
fire_long = [fire[0][0] for fire in my_fires['Fires']]
fire_lat = [fire[0][1] for fire in my_fires['Fires']]
```
