## Fast Model Exploration

This notebook is to experiment with many different classification models to quickly determine which type of model yields useful results for our fires dataset. We have yet to pair additional datasources or do significant feature engineering.



```python
import numpy as np, pandas as pd
import warnings
import hdbscan
warnings.filterwarnings('ignore')
# warnings.filterwarnings('DataConversionWarning')
```

```python
#  I will handle the time features of the dataframe in an external function that I can build into the api
# I'll just read out pickles directly into this funciton
def process_live_data_and_cluster(original_df):
    """
    Pre processes live data to match pipeline expectations.
    """
    df = original_df.copy()
    df['acq_date'] = pd.to_datetime(df['acq_date'])
    df['satellite'] = df['satellite'].replace({'T':'Terra', 'A': 'Aqua'})
    df['latitude'] = df['latitude'].round(2)
    df['longitude'] = df['longitude'].round(2) 
    
    df['month'] = df['acq_date'].dt.month
    df['week'] = df['acq_date'].dt.weekofyear
    df.drop(columns=['acq_date', 'acq_time'], inplace=True)
    
    #add binary cluster label
    clusterer = hdbscan.HDBSCAN(min_cluster_size=50, min_samples=20, gen_min_span_tree=True)
    df['clusters'] = clusterer.fit(df[['latitude','longitude']]).labels_
    df['clusters'] = df['clusters'].apply(lambda x: False if x == -1 else True)
    
    
    return df



X_train = process_live_data_and_cluster(pd.read_pickle('~/datascience/burn notice/Data-Science/Data/Cleaned_Modus_Historical/X_train_small.pkl'))
y_train = (pd.read_pickle('~/datascience/burn notice/Data-Science/Data/Cleaned_Modus_Historical/y_train_small.pkl'))
X_test = process_live_data_and_cluster(pd.read_pickle('~/datascience/burn notice/Data-Science/Data/Cleaned_Modus_Historical/X_test_small.pkl'))
y_test = (pd.read_pickle('~/datascience/burn notice/Data-Science/Data/Cleaned_Modus_Historical/y_test_small.pkl'))
```

```python
X_train.head()
```

```python
#custom transformers, the dense transformer is for the GuassianNB classifier
from sklearn.base import TransformerMixin
class DenseTransformer(TransformerMixin):

    def fit(self, X, y=None, **fit_params):
        return self

    def transform(self, X, y=None, **fit_params):
        return X.todense()
```

```python
# Now I'll set up pipelines
# scikit-learn pipelines
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer

# feature processing
from sklearn.preprocessing import StandardScaler, OneHotEncoder

    
# deals with one hot encoding
ohe_features = ['satellite', 'daynight', 'month', 'week', 'clusters']
# scaler
scale_features = ['brightness', 'track', 'scan', 'bright_t31', 'frp', 'confidence']

column_trans = make_column_transformer(
    (OneHotEncoder(), ohe_features),
    (StandardScaler(), scale_features), remainder='drop')
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# we create classification pipelines
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from xgboost import XGBClassifier 
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC






random_state = 314

pipelines ={
    'sgd' : make_pipeline(column_trans, SGDClassifier(random_state = random_state)),
    'knn' : make_pipeline(column_trans, KNeighborsClassifier()),
    'dtc' : make_pipeline(column_trans, DecisionTreeClassifier(random_state = random_state)),
    'svc' : make_pipeline(column_trans, LinearSVC(random_state = random_state)),
    'gnb' : make_pipeline(column_trans, DenseTransformer(), GaussianNB()),
    'rfc' : make_pipeline(column_trans, RandomForestClassifier(n_estimators=100, random_state = random_state )),
    'gbc' : make_pipeline(column_trans, GradientBoostingClassifier(random_state = random_state)),
    'ada' : make_pipeline(column_trans, AdaBoostClassifier(random_state = random_state)),
    'bag' : make_pipeline(column_trans, BaggingClassifier(random_state = random_state)),
    'etc' : make_pipeline(column_trans, ExtraTreesClassifier(random_state = random_state)),
    'lgr' : make_pipeline(column_trans, LogisticRegression(random_state = random_state)),
    'xgb' : make_pipeline(column_trans, XGBClassifier(random_state = random_state))
}

new_models = ['gnb', 'sgd', 'knn', 'dtc', 'svc']


for key, value in pipelines.items():
    value.fit(X_train, y_train)
    print(key + ' score: ', value.score(X_test, y_test))
```

```python
# suspiciously high accuracy, let's check ROC and f1
from sklearn.metrics import f1_score, roc_auc_score, recall_score

results = []

for key, value in pipelines.items():
    
    y_preds = value.predict(X_test)
    results.append({'model' : key + ' test', 
                    'accuracy' : value.score(X_test, y_test),
                    'f1' : f1_score(y_test, y_preds),
                    'ROC' : roc_auc_score(y_test, y_preds),
                    'recall' : recall_score(y_test, y_preds)
                   })
r_df = pd.DataFrame(results)
```

```python
r_df.sort_values(by= ['recall','f1', 'ROC'], ascending=False)[:5]
```

```python
# validation imports
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

def normalized_confusion_matrix(y_test, y_pred, title):
    
    cm = confusion_matrix(y_test, y_pred) 
    #normalize cm
    array = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    df_cm = pd.DataFrame(array, range(2),
                      range(2))
    #plt.figure(figsize = (10,7))
    sn.set(font_scale=1.4)#for label size
    ax = sn.heatmap(df_cm, annot=True,annot_kws={"size": 16})#.set_title('Normalized confusion matrix')# font size
    ax.set(xlabel='Predicted Fires', ylabel='True Fires', title = title)
    plt.show();
```

```python
# gnb is doing amazing, I'm going to visuallize this
normalized_confusion_matrix(y_test, pipelines['gnb'].predict(X_test), 'Gaussian Naive Bayes normalized confusion matrix')
```

```python
normalized_confusion_matrix(y_test, pipelines['sgd'].predict(X_test), 'sgd normalized confusion matrix')
```

```python
normalized_confusion_matrix(y_test, pipelines['lgr'].predict(X_test), 'lgr normalized confusion matrix')
```

```python
# now I'll evaluate a fire prediction metric of confidence > 90 vs the model

def predict_90(row):
    
    if row['confidence'] > 90:
        return 1
    else:
        return 0

y_pred_c90 = X_test.apply(lambda row : predict_90(row), axis=1)

normalized_confusion_matrix(y_test, y_pred_c90, 'Normalized confusion matrix for y_pred>90')
```
